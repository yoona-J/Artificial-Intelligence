```python
from sklearn import datasets
from sklearn import svm
import plotly.express as px

# iris 데이터 불러오기, 150개의 데이터(d.data)가 3개의 레이블(d.target)로 나뉘어 있다.
d = datasets.load_iris()
# iris 데이터 출력
print(d.DESCR)

# 샘플 순서대로 출력
for i in range(0, len(d.data)):
  print (i+1, d.data[i], d.target[i])

# SVM 중 SVC 객체 생성
s = svm.SVC(gamma = 0.1, C = 10)
# iris 데이터로 학습(fit) 
s.fit(d.data, d.target)

# 101, 51번째 데이터를 변형하여 새로운 데이터 생성
new_d = [[6.4, 3.2, 6.0, 2.5], [7.1, 3.1, 4.7, 1.35]]

# new_d 데이터를 이용해 예측(predict)
res = s.predict(new_d)

print('새로운 2개 샘플 부류는: ', res)

# 3차원 구성
df = px.data.iris()
fig = px.scatter_3d(df, x = 'petal_length', y = 'sepal_width', z = 'petal_width', color = 'species')
fig.show()
```
